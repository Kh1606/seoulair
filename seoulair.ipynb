```python
import pandas as pd
import numpy as np

# Load your CSV file
df = pd.read_csv('/Users/saidev/Downloads/tayyor2/2022.csv')

# List of regions you want to add
regions = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '노원구', '도봉구', '동대문구', '동작구', '마포구',
 '금천구', '서초구', '서대문구', '중구', '종로구', '은평구', '용산구', '중랑구', '평균', '양천구', '송파구', '성북구',
 '성동구', '영등포구']  # Add as many regions as you need

# Custom ranges for each column (assuming columns are numeric)
column_ranges = {
    'temp': (-2, 2),  # Adjust range for each column
    'precipitation(mm)': (0, 1),
    'wind speed(m/s)': (0, 1.5),
    'humidity': (-4, 4),
    'vapor_pressure': (-0.5, 0.5),
    'dew_point': (-0.5, 0.5),
    'local_pressure': (-0.5, 0.5),
    'sea_level_pressure': (-0.5, 0.5),
    'sunshine': (0, 0.5),
    'ilsa(MJ/m2)': (0, 0.5),
    'snow_cover': (0, 0.5),
    'total_cloud_volume(10min)': (0, 1),
    'lower/middle_cloud_volume(10min)': (0, 1),
    'minimum_cloud_height(100m)': (-5, 5),
    'ground_temp': (-2, 2),
}

# Create an empty list to store all new rows
expanded_rows = []

# Function to apply random variation to a DataFrame's columns and round to match original precision
def apply_random_variation(df, column_ranges):
    df_modified = df.copy()
    for col, (low, high) in column_ranges.items():
        if col in df.columns:
            # Apply random variation
            variation = np.random.uniform(low, high, df.shape[0])
            
            # Round to match the decimal precision of the original column
            decimal_precision = df[col].apply(lambda x: len(str(x).split('.')[1]) if '.' in str(x) else 0).max()
            df_modified[col] = (df[col] + variation).round(decimal_precision)
    
    # For 'minimum_cloud_height(100m)', round to nearest natural number but keep ".0"
    if 'minimum_cloud_height(100m)' in df_modified.columns:
        df_modified['minimum_cloud_height(100m)'] = df_modified['minimum_cloud_height(100m)'].apply(
            lambda x: f"{int(round(x))}.0" if not np.isnan(x) else np.nan
        )
    
    return df_modified

# Group the data by date to iterate over each day
grouped_by_date = df.groupby('date')

# Iterate over each day
for date, group in grouped_by_date:
    for region in regions:
        # Apply random variation based on the defined ranges
        df_region = apply_random_variation(group, column_ranges)
        df_region['region'] = region  # Add region column with current region name
        df_region['date'] = group['date'].values  # Ensure date remains the same
        expanded_rows.append(df_region)  # Append region's data for the day

# Concatenate all the rows into a single DataFrame
expanded_df = pd.concat(expanded_rows, ignore_index=True)

# Reorder columns: Date first, Region second
expanded_df = expanded_df[['date', 'region'] + [col for col in expanded_df.columns if col not in ['date', 'region']]]

# Save the new DataFrame back to a CSV file
output_file_path = '/Users/saidev/Downloads/tayyor2/2022_zayebal.csv'
expanded_df.to_csv(output_file_path, index=False, float_format="%.1f")

print(f"Data expanded with regions and saved to {output_file_path}")
```

```python
import pandas as pd

# Load the CSV file
file_path = '/Users/saidev/Downloads/tayyor2/2022_wwww.csv'
df = pd.read_csv(file_path)

# Convert the 'date' column to datetime type for sorting
df['date'] = pd.to_datetime(df['date'])

# Extract the date part (without time) for grouping
df['day'] = df['date'].dt.date

# List of all regions you want to process
regions = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '노원구', '도봉구', '동대문구', '동작구', '마포구',
           '금천구', '서초구', '서대문구', '중구', '종로구', '은평구', '용산구', '중랑구', '평균', '양천구', '송파구', '성북구',
           '성동구', '영등포구']

# Sort the dataframe by 'date' and 'region' (ensure region is sorted in the order of regions list)
df_sorted = df.sort_values(by=['date', 'region'], key=lambda col: col.map({region: i for i, region in enumerate(regions)}))

# Create an empty dataframe to store the results
df_reordered = pd.DataFrame()

# Get unique days in the dataset
unique_days = sorted(df['day'].unique())

# Iterate through the days and append data for each region in the specified order for each day
for day in unique_days:
    for region in regions:
        region_df = df_sorted[(df_sorted['day'] == day) & (df_sorted['region'] == region)]
        df_reordered = pd.concat([df_reordered, region_df])

# Drop the extra 'day' column used for grouping
df_reordered.drop(columns=['day'], inplace=True)

# Specify the directory where you want to save the result file
output_directory = '/Users/saidev/Downloads/tayyor2/2022_test.csv'

# Save the reordered dataframe to the specified directory
df_reordered.to_csv(output_directory, index=False)

print(f"Data reordered by day and regions and saved to {output_directory}")
```
